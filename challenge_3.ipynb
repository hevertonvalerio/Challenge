```python
import yfinance as yf
import pandas as pd

# Lista de tickers das ações que você deseja calcular o beta
stock_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN']  # Exemplo de ações da Apple, Microsoft, Google e Amazon
index_ticker = '^GSPC'  # Índice de referência, exemplo: S&P 500

# Defina as datas de início e fim
start_date = '2023-01-01'
end_date = '2024-01-01'

# Baixar os dados históricos de preços do índice de referência
index_data = yf.download(index_ticker, start=start_date, end=end_date, interval='1d')
index_returns = index_data['Adj Close'].pct_change().dropna()

# Criar um dicionário para armazenar os betas
betas = {}

# Loop pelos tickers das ações
for stock_ticker in stock_tickers:
    # Baixar os dados históricos de preços da ação
    stock_data = yf.download(stock_ticker, start=start_date, end=end_date, interval='1d')
    stock_returns = stock_data['Adj Close'].pct_change().dropna()

    # Alinhar as datas dos retornos para garantir que sejam comparáveis
    returns = pd.DataFrame({'stock': stock_returns, 'index': index_returns}).dropna()

    # Calcular a covariância entre a ação e o índice e a variância do índice
    covariance = returns['stock'].cov(returns['index'])
    variance = returns['index'].var()

    # Calcular o beta
    beta = covariance / variance

    # Armazenar o beta no dicionário
    betas[stock_ticker] = beta

# Exibir os betas de cada ação
for ticker, beta_value in betas.items():
    print(f'O beta da ação {ticker} em relação ao índice {index_ticker} é: {beta_value}')
```

```python
import yfinance as yf
import pandas as pd

# Lista de tickers de empresas do setor de tecnologia (exemplo)
tech_tickers = [
    'AAPL',  # Apple Inc.
    'MSFT',  # Microsoft Corporation
    'GOOGL',  # Alphabet Inc. (Google)
    'AMZN',  # Amazon.com, Inc.
    'FB',    # Meta Platforms, Inc. (Facebook)
    'TSLA',  # Tesla, Inc.
    'NVDA',  # NVIDIA Corporation
    'ADBE',  # Adobe Inc.
    'CSCO',  # Cisco Systems, Inc.
    'INTC'   # Intel Corporation
]

# Dicionário para armazenar as informações
data = {}

# Loop através dos tickers para obter informações
for ticker in tech_tickers:
    try:
        stock_info = yf.Ticker(ticker).info
        data[ticker] = {
            'Company Name': stock_info.get('longName', 'N/A'),
            'Sector': stock_info.get('sector', 'N/A'),
            'Market Cap': stock_info.get('marketCap', 'N/A'),
            'Current Price': stock_info.get('currentPrice', 'N/A')
        }
    except Exception as e:
        print(f"Erro ao obter dados para {ticker}: {e}")

# Criar um DataFrame a partir dos dados
df = pd.DataFrame.from_dict(data, orient='index')

# Exibir o DataFrame
print(df)
```
